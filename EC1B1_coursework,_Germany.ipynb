```python
%matplotlib inline
import math
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# importing germany dataset 
germany_df = pd.read_excel('germany_data.xlsx', skiprows=2, index_col=0)
germany_df.index = pd.to_datetime(germany_df.index)
germany_df.columns = ['Industrial Production', 'germany/usd', 'germany_reserves', 'germany_cpi']

# importing usa dataset
usa_df = pd.read_excel('usa_data.xlsx', skiprows=2, index_col=0)
usa_df.index = pd.to_datetime(usa_df.index)
usa_df.columns = ['usa_reserves', 'usa_cpi']

# merge the datasets
joint = pd.concat([germany_df, usa_df], axis=1)

# log(realXR) = log(nominalXR) + log(foreign price) - log(domestic price)
# change in log(realXR) = change in log(nominalXR) + foreign inflation - domestic inflation

# FOR GERMANY:

# (i) log exchange rate
germany_log_XR = np.log(joint['germany/usd'])

# (ii) The difference in the log exchange rate versus the previous month
germany_change_log_XR = germany_log_XR.diff(1)

# (iii) The monthly inflation rate
germany_log_cpi = np.log(joint['germany_cpi'])
germany_inflation = germany_log_cpi.diff(1)

# (iv) Log industrial production
germany_log_industrial_production = np.log(joint['Industrial Production'])

# (v) Monthly growth in industrial production
germany_MONTHLY_growth_industrial_production = germany_log_industrial_production.diff(1)

# (vi) Growth in industrial production versus 12 months ago
germany_YEARLY_growth_industrial_production = germany_log_industrial_production.diff(12)

# (vii) An index of the value of international reserves
germany_reserve_index = np.multiply((np.divide(joint['germany_reserves'], joint['germany_reserves'][0])), 100)
germany_df_clean = pd.concat([germany_log_XR, germany_change_log_XR, germany_inflation, germany_log_industrial_production, germany_MONTHLY_growth_industrial_production, germany_YEARLY_growth_industrial_production, germany_reserve_index], axis=1)
germany_df_clean.columns = ["germany log x/r", 'germany change in XR', 'germany inflation', 'germany log production', 'germany monthly growth in production', 'germany yearly growth in production', 'germany_reserve_index']

#FOR USA

# (i) The monthly inflation rate
usa_log_cpi = np.log(joint['usa_cpi'])
usa_inflation = usa_log_cpi.diff(1)

# (ii) An index of the value of international reserves
usa_reserve_index = np.multiply((np.divide(joint['usa_reserves'], joint['usa_reserves'][0])), 100)
usa_df_clean = pd.concat([usa_inflation, usa_reserve_index], axis=1)
usa_df_clean.columns = ['usa inflation', 'usa reserve index']

# cleaning the data
combined_df = pd.concat([usa_df_clean, germany_df_clean], axis=1)

# identifying outliers and setting them to missing
def outliers(dataframe):
    q1 = dataframe.quantile(0.2)
    q3 = dataframe.quantile(0.8)
    iqr = q3 - q1
    outliers_df = ((dataframe < (q1 - 2*iqr)) | (dataframe > (q3 + 2*iqr)))
    dataframe[outliers_df] = float("nan") 
    return dataframe, outliers_df

combined_df, outliers = outliers(combined_df)
print("\nMissing values:")
print(combined_df.isnull().sum())
print("\nOutliers:")
print(outliers.sum())

# interpolating missing values
for col in combined_df.columns:
    combined_df[col] = combined_df[col].interpolate(method='linear')

print("\nMissing values after interpolation:")
print(combined_df.isnull().sum())

germany_change_in_real_xr = combined_df["germany change in XR"] + combined_df['germany inflation'] - combined_df["usa inflation"]
df =  pd.concat([combined_df, germany_change_in_real_xr.rename("germany change in real XR")], axis = 1)
pd.set_option('display.max_rows',400)
display(df)
df.shape

print("Germany left the Bretton Woods system on 1/5/1971, this is indicated by the dashed red line")

# Germany nominal XR growth vs USD
fig, ax = plt.subplots(figsize=(20, 7))
date_format = mdates.DateFormatter('%d/%m/%Y')
ax.xaxis.set_major_formatter(date_format)
ax.set_xlabel('Date', fontsize='large')
ax.set_ylabel('Nominal Exchange Rate Log Change', fontsize='large')
ax.plot(df.index[df.index.year<=1991], df['germany change in XR'][df.index.year<=1991], color='green')
ax.set_xlim(dt.datetime(1960, 1, 1), dt.datetime(1990, 12, 1))
year_locator = mdates.YearLocator(2)
ax.xaxis.set_major_locator(year_locator)
ax.xaxis.grid(True)
plt.title('Germany nominal XR growth vs USD\n1/1/1960-31/12/1990', fontsize='x-large')
plt.axvline(dt.datetime(1973,3,1), color='red', linestyle='--')
plt.show()

# Germany real XR growth vs USD
fig, ax = plt.subplots(figsize=(20, 7))
date_format = mdates.DateFormatter('%d/%m/%Y')
ax.xaxis.set_major_formatter(date_format)
ax.set_xlabel('Date', fontsize='large')
ax.set_ylabel('Real Exchange Rate Log Change', fontsize='large')
ax.plot(df.index[df.index.year<=1991], df['germany change in real XR'][df.index.year<=1991], color='green')
ax.set_xlim(dt.datetime(1960, 1, 1), dt.datetime(1990, 12, 1))
year_locator = mdates.YearLocator(2)
ax.xaxis.set_major_locator(year_locator)
ax.xaxis.grid(True)
plt.title('Germany real XR growth vs USD\n1/1/1960-31/12/1990', fontsize='x-large')
plt.axvline(dt.datetime(1973,3,1), color='red', linestyle='--')
plt.show()

# Monthly inflation
print("\n")
fig, ax = plt.subplots(figsize=(20, 7))
date_format = mdates.DateFormatter('%d/%m/%Y')
ax.xaxis.set_major_formatter(date_format)
ax.plot(df.index, df['germany inflation']*100, color='green')
ax.set_xlim(dt.datetime(1960, 1, 1), dt.datetime(1990, 12, 1))
ax.axvline(dt.datetime(1973,3,1), color='red', linestyle='--')
ax.set_xlabel('Date', fontsize='large')
ax.set_ylabel('Monthly Inflation', fontsize='large')
ax.set_title('Germany Monthly Inflation (%)\n1/1/1960-31/12/1990', fontsize='x-large')
year_locator = mdates.YearLocator(2)
ax.xaxis.set_major_locator(year_locator)
ax.xaxis.grid(True)
plt.show()

# Monthly growth of industrial production 
print("\n")
fig, ax = plt.subplots(figsize=(20, 7))
date_format = mdates.DateFormatter('%d/%m/%Y')
ax.xaxis.set_major_formatter(date_format)
ax.plot(df.index, df['germany monthly growth in production'], color='green')
ax.set_xlim(dt.datetime(1960, 1, 1), dt.datetime(1990, 12, 1))
ax.axvline(dt.datetime(1973,3,1), color='red', linestyle='--')
ax.set_xlabel('Date', fontsize='large')
ax.set_ylabel('Log change in monthly industrial production growth rate (%)', fontsize='large')
ax.set_title('Germany Monthly Growth in Industrial Production\n1/1/1960-31/12/1990', fontsize='x-large')
year_locator = mdates.YearLocator(2)
ax.xaxis.set_major_locator(year_locator)
ax.xaxis.grid(True)
plt.show()

# Annual growth of industrial production 
print("\n")
fig, ax = plt.subplots(figsize=(20, 7))
date_format = mdates.DateFormatter('%d/%m/%Y')
ax.xaxis.set_major_formatter(date_format)
ax.plot(df.index, df['germany yearly growth in production'], color='green')
ax.set_xlim(dt.datetime(1960, 1, 1), dt.datetime(1990, 12, 1))
ax.axvline(dt.datetime(1973,3,1), color='red', linestyle='--')
ax.set_xlabel('Date', fontsize='large')
ax.set_ylabel('Log change in annual industrial production growth rate (%)', fontsize='large')
ax.set_title('Germany Annual Growth in Industrial Production\n1/1/1960-31/12/1990', fontsize='x-large')
year_locator = mdates.YearLocator(2)
ax.xaxis.set_major_locator(year_locator)
ax.xaxis.grid(True)
plt.show()

#Index of International Reserves
print("\n")
fig, ax = plt.subplots(figsize=(20, 7))
date_format = mdates.DateFormatter('%d/%m/%Y')
ax.xaxis.set_major_formatter(date_format)
ax.set_xlabel('Date', fontsize='large')
ax.set_xlim(dt.datetime(1960,1,1), dt.datetime(1973,3,1))
ax.set_ylim(0, 400)
ax.plot(df.index, df['germany_reserve_index'], label="Germany", color="green")
ax.plot(df.index, df['usa reserve index'], label="USA", color="blue")
ax.set_title('International Reserves\n1/1/1960-1/5/1971', fontsize='x-large')
ax.set_ylabel('Reserve Index Value', fontsize='large')
ax.legend()
year_locator = mdates.YearLocator(2)
ax.xaxis.set_major_locator(year_locator)
ax.xaxis.grid(True)
plt.show()

# CALCULATING STANDARD DEVIATIONS

# SD before leaving Bretton Woods
nominal_XR_std1 = df.loc[df.index <= dt.datetime(1972, 4, 1), 'germany change in XR'].std()
real_XR_std1 = df.loc[df.index <= dt.datetime(1972, 4, 1), 'germany change in real XR'].std()
print("\nNominal XR standard deviation before leaving Bretton Woods:", nominal_XR_std1)
print("Real XR standard deviation before leaving Bretton Woods:", real_XR_std1)

# SD after leaving Bretton Woods
nominal_XR_std2 = df.loc[df.index >= dt.datetime(1974, 2, 1), 'germany change in XR'].std()
real_XR_std2 = df.loc[df.index >= dt.datetime(1974, 2, 1), 'germany change in real XR'].std()
print("\nNominal XR standard deviation after leaving Bretton Woods:", nominal_XR_std2)
print("Real XR standard deviation after leaving Bretton Woods:", real_XR_std2)

# SD Monthly inflation Germany
inflation_before_std = df.loc[df.index <= dt.datetime(1972, 4, 1), 'germany inflation'].std()
print("\ninflation standard deviation before leaving Bretton Woods:", inflation_before_std)
inflation_after_std = df.loc[df.index >= dt.datetime(1974, 2, 1), 'germany inflation'].std()
print("inflation standard deviation after leaving Bretton Woods:", inflation_after_std)

# SD Monthly inflation Germany - USA
inflation_before_std = (df.loc[df.index <= dt.datetime(1972, 4, 1), 'germany inflation'] - df.loc[df.index <= dt.datetime(1972, 4, 1), 'usa inflation']).std()
print("\nstandard deviation of difference in inflation before leaving Bretton Woods:", inflation_before_std)
inflation_after_std = (df.loc[df.index >= dt.datetime(1974, 2, 1), 'germany inflation'] - df.loc[df.index >= dt.datetime(1974, 2, 1), 'usa inflation']).std()
print("standard deviation of difference in inflation after leaving Bretton Woods:", inflation_after_std)

# SD of 12-month industrial production growth
production_std = df.loc[df.index <= dt.datetime(1972, 4, 1), 'germany monthly growth in production'].std()
print("\n12-month industrial production growth standard deviation before leaving Bretton Woods:", production_std)
production_after_std = df.loc[df.index >= dt.datetime(1974, 2, 1), 'germany monthly growth in production'].std()
print("12-month industrial production growth standard deviation after leaving Bretton Woods:", production_after_std)

# Ratio before/after
ratio_nominal_XR = nominal_XR_std1/nominal_XR_std2
print("\nnormal XR ratio before/after =", ratio_nominal_XR)
ratio_real_XR = real_XR_std1/real_XR_std2
print("\nreal XR ratio before/after =", ratio_real_XR)
ratio_monthly_inflation = inflation_before_std/inflation_after_std
print("\nmonthly inflation ratio before/after =", ratio_monthly_inflation)
ratio_production = production_std/production_after_std
print("\nproduction ratio before/after =", ratio_production)
```
